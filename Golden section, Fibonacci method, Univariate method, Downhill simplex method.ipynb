```python
def g_lambda(a,b):
    return a+(1-alpha)*(b-a)

def g_mu(a,b):
    return a+alpha*(b-a)

def goldSearch(a,b,lambda_k,mu_k,function,k = 1,print_a_b = True):
    if print_a_b:
        print "a = %f , b = %f" %(a,b)
    # step1 
    if (b - a) < l:
        return ((a+b)/2.,k)
    
    if function(lambda_k) > function(mu_k):
        # step2 
        a = lambda_k
        b = b
        lambda_k = mu_k
        mu_k = g_mu(a,b)
        k = k+1
        return goldSearch(a,b,lambda_k,mu_k,function,k,print_a_b)
    
    elif function(lambda_k) <= function(mu_k):
        # step3
        a = a
        b = mu_k
        mu_k = lambda_k
        lambda_k = g_lambda(a,b)
        k = k+1
        return goldSearch(a,b,lambda_k,mu_k,function,k,print_a_b)
```

### Golden section method
  minimum points x = 1.26  
  iteration n = 26 

```python
def funcname(x):
    return (x-1)*(x-1.5)*math.sin(x)
l = 10**-5
alpha = (math.sqrt(5)-1)/2.
a = 0
b = 1.5
goldSearch(a,b,g_lambda(a,b),g_mu(a,b),funcname)
```

  a,b = -5,5

```python
x_y_list = []

def xfuncname(x):
    return 2*(x**2)-2*x*(y**2)-2*x+(y**4)

def yfuncname(y):
    return 2*(x**2)-2*x*(y**2)-2*x+(y**4)

# –5 < x < 5
# –5 < y < 5
e = 0.000001
a = -5
b = 5

# 先從 x 開始
y = -5
x1ast1 = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),xfuncname,print_a_b = False)[0]
x = x1ast1
x_y_list.append((x,y))
y1ast1 = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),yfuncname,print_a_b = False)[0]
y = y1ast1
x_y_list.append((x,y))
x = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),xfuncname,print_a_b = False)[0]
x_y_list.append((x,y))
y = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),yfuncname,print_a_b = False)[0]
x_y_list.append((x,y))

while abs(x-x1ast1) > e and abs(y-y1ast1) > e:
    x1ast1 = x
    y1ast1 = y
    x = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),xfuncname,print_a_b = False)[0]
    x_y_list.append((x,y))
    y = goldSearch(a,b,g_lambda(a,b),g_mu(a,b),yfuncname,print_a_b = False)[0]
    x_y_list.append((x,y))

print x
print y
```

```python
def xmax_funcname(x_list):
    x_tuple = [(x,funcname(x)) for x in x_list]
    return max(x_tuple,key=lambda item:item[1])[0]

def secondMax_funcname(x_list):
    x_tuple = [(x,funcname(x)) for x in x_list]
    max_value = max(x_tuple,key=lambda item:item[1])[1]
    for i,x in enumerate(x_tuple):
        if x[1] == max_value:
            x_tuple.pop(i)
            break
    return max(x_tuple,key=lambda item:item[1])[0]

def xmin_funcname(x_list):
    x_tuple = [(x,funcname(x)) for x in x_list]
    return min(x_tuple,key=lambda item:item[1])[0]

center_list = []
def DownhillSimplexMethod(x_list, n=1):
    center = (sum(map(lambda x: x[0],x_list))/3.,sum(map(lambda x: x[1],x_list))/3.)
    center_list.append(center)
    xmax = xmax_funcname(x_list)
    xmin = xmin_funcname(x_list)
    if math.sqrt(sum((xmax - xmin) * (xmax - xmin))) > e:
        xa = (np.array((sum(map(lambda x: x[0],x_list)),sum(map(lambda x: x[1],x_list))))-xmax)/2.
        # step2
        xr = xa + alpha*(xa-xmax)
        if funcname(xmin) > funcname(xr):
            xe = xa + gamma*(xr-xa)
            # step3
            if funcname(xr) > funcname(xe):
                for i,v in enumerate(x_list):
                    if list(v) == list(xmax):
                        x_list.pop(i)
                        break
                xmax = xe
                x_list.append(xmax)
            else:
                for i,v in enumerate(x_list):
                    if list(v) == list(xmax):
                        x_list.pop(i)
                        break
                xmax = xr
                x_list.append(xmax)
            n = n+1
            return DownhillSimplexMethod(x_list, n)
        # step 4
        else :
            if funcname(secondMax_funcname(x_list)) >= funcname(xr):
                for i,v in enumerate(x_list):
                    if list(v) == list(xmax):
                        x_list.pop(i)
                        break
                xmax = xr
                x_list.append(xmax)
                n = n+1
                return DownhillSimplexMethod(x_list, n)
            # step 5
            else:
                xp = xmin_funcname([xr,xmax])
                xc = xa + beta*(xp-xa)
                if funcname(xc) > funcname(xp):
                    x_list = map(lambda x: x+((xmin-x)/2.),x_list)
                    n = n+1
                    return DownhillSimplexMethod(x_list, n)
                else:
                    for i,v in enumerate(x_list):
                        if list(v) == list(xmax):
                            x_list.pop(i)
                            break
                    xmax = xc
                    x_list.append(xmax)
                    n = n+1
                    return DownhillSimplexMethod(x_list, n)
    return x_list,n
```

```python
def funcname(xn):
    x = xn[0]
    y = xn[1]
    return 2*(x**2)-2*x*(y**2)-2*x+(y**4)

e = 0.0001
alpha = 1.
beta = 0.5
gamma = 2.
x1 = np.array([7,0])
x2 = np.array([-2,2])
x3 = np.array([1,1])
x_list = [x1,x2,x3]
```
